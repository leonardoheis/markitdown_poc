```python
from markitdown import MarkItDown

md = MarkItDown()
```

```python
result_pdf = md.convert("data/Leonardo_Heis_2026.pdf")
print(result_pdf.text_content)
```

```python
result_pdf = md.convert("data/convenio_1_2013.pdf")
print(result_pdf.text_content)
```

```python
result_xl = md.convert("data/DIA_A_Grupos_ACTUALIZADOS.xlsx")
print(result_xl.text_content)
transcript_text = result_xl.text_content
```

```python
message = {
    "role": "user",
    "description": "You are an AI assistant helping users to summarize the provided content. Please provide summary in 5 bullet points.",
    "content": transcript_text,
}
```

```python
import ollama

# Use ollama.chat function to send the transcript and retrieve the summary
response = ollama.chat(model="llama3.2", messages=[message])
print(response)
```
